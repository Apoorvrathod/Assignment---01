```python
#Q1. What is Flask Framework? What are the advantages of Flask Framework?
"""Flask Framework is a lightweight web framework written in Python. It is designed to be simple and easy to use, making
it a popular choice for developing web applications. Flask follows the Model-View-Controller (MVC) architectural pattern,
although it is often categorized as a microframework because it does not require particular tools or libraries."""

#Advantages of Flask Framework include:

"""Lightweight and minimalistic: Flask has a small core and only provides the essential features needed to build a web 
application. It allows developers to add extensions and libraries as per their specific requirements, keeping the codebase
minimal and easy to maintain.
Easy to get started: Flask has a simple and intuitive syntax, making it easy for beginners to start building web applications.
The learning curve is relatively low compared to other web frameworks.
Flexible and extensible: Flask allows developers to choose and integrate various libraries and tools based on their 
project needs. It provides a wide range of extensions for database integration, form validation, authentication, and more.
Well-documented: Flask has comprehensive and well-organized documentation, along with a large and active community. This 
makes it easy to find solutions to common problems and get support when needed."""
```

```python
#Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/first print")
def hello_world():
    return "<h1>Hello, World!!</h1>"
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q3. What is App routing in Flask? Why do we use app routes?
"""App routing in Flask refers to the mechanism of associating URLs or routes with specific functions in your Flask application.
It allows you to define different URL patterns and map them to corresponding Python functions, also known as view functions.
"""
"""URL handling: App routes define the URLs that users can visit to interact with different parts of your application. By
defining routes, you can specify which function should be executed when a particular URL is accessed.

Endpoint mapping: Each route in Flask is associated with an endpoint, which is a unique name assigned to the route. 
Endpoints are used to generate URLs for specific routes, and they allow for flexibility when changing the URL structure of 
your application without affecting the code that references the URLs.

Request handling: When a user visits a URL associated with a specific route, the corresponding view function is executed.
View functions can retrieve data from requests, perform operations, and return responses to the client. App routes help in
organizing and structuring the request handling process.

Dynamic routing: Flask supports dynamic routes, which allow parts of the URL to be dynamic and passed as variables to view 
functions."""
```

```python
"""Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999"""
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
```
